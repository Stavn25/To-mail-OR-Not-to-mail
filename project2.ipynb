```python
target='V86'
```

```python
bd_train[target].value_counts()
```

```python
x_train=bd_train.drop(target,1)
y_train=bd_train[target]
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(fit_intercept=True)
```

```python
params={'penalty':['l1','l2'],
       'C':[.0001,.0005,.001,.005,.01,.05,.1,1,2,5],
       'class_weight':['balanced',None]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gs=GridSearchCV(model,cv=10,param_grid=params,n_jobs=-1,verbose=5,scoring='roc_auc')
```

```python
gs.fit(x_train,y_train)
```

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.5f} (std: {1:.5f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```

```python

report(gs.cv_results_,5)
```

```python
gs.best_estimator_
```

```python
train_score=gs.best_estimator_.predict_proba(x_train)[:,1]
```

```python
real=y_train
```

```python
cutoffs=np.linspace(0.001,0.999,999)
```

```python
from sklearn.metrics import fbeta_score
```

```python
fbetas=[]
```

```python
for cutoff in cutoffs:
    predicted=(train_score>cutoff).astype(int)
    fbetas.append(fbeta_score(y_train,predicted,beta=2))
```

```python
my_cutoff=cutoffs[fbetas==max(fbetas)]
```

```python
predictions=(gs.predict_proba(bd_test)[:,1]>my_cutoff).astype(int)
```

```python
pd.Series(predictions).value_counts()
```

```python
submissions=pd.DataFrame({'V86':predictions})
```

```python
# submissions.to_csv('sample_submission_proj2_01.csv',index=False)
```

```python
KS_all=[]

for cutoff in cutoffs:
    
    predicted=(train_score>cutoff).astype(int)

    TP=((predicted==1) & (real==1)).sum()
    TN=((predicted==0) & (real==0)).sum()
    FP=((predicted==1) & (real==0)).sum()
    FN=((predicted==0) & (real==1)).sum()
    
    P=TP+FN
    N=TN+FP
    
      
    KS=(TP/P)-(FP/N)
       
    KS_all.append(KS)
    
```

```python

list(zip(cutoffs,KS_all))
```

```python
mycutoff=cutoffs[KS_all==max(KS_all)]
mycutoff
```
